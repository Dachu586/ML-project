```python


def classify_mobile_price(brand, ram, storage, camera, battery, screen_size):
    # Simple classification logic (you can make this more sophisticated)
    score = 0

    if brand.lower() in ['apple', 'samsung', 'google']:
        score += 30
    elif brand.lower() in ['oneplus', 'xiaomi', 'huawei']:
        score += 20
    else:
        score += 10

    score += ram * 5
    score += storage // 32 * 10
    score += camera // 12 * 15
    score += battery // 1000 * 10
    score += screen_size * 5

    if score < 50:
        return "Budget (< $200)"
    elif score < 100:
        return "Mid-range ($200 - $500)"
    elif score < 150:
        return "High-end ($500 - $800)"
    else:
        return "Premium (> $800)"

# Create the Gradio interface
iface = gr.Interface(
    fn=classify_mobile_price,
    inputs=[
        gr.Textbox(label="Brand"),
        gr.Slider(1, 16, step=1, label="RAM (GB)"),
        gr.Slider(16, 512, step=16, label="Storage (GB)"),
        gr.Slider(8, 108, step=1, label="Main Camera (MP)"),
        gr.Slider(2000, 6000, step=100, label="Battery Capacity (mAh)"),
        gr.Slider(4.0, 7.0, step=0.1, label="Screen Size (inches)")
    ],
    outputs=gr.Textbox(label="Price Category"),
    title="Mobile Price Classification",
    description="Enter the specifications of a mobile phone to get its price category."
)

# Launch the interface
iface.launch()
```
